### Difference between a local and global Python environment

By default, Python is installed globally, meaning that all users and applications use the same version and, more importantly, the same version of packages (libraries). As we might have some applications that run only with a specific version of a library, it is hard and sometimes impossible to resolve version conflicts system-wide.

Therefore, it is possible to create *environments*, of which you can think of as encapsulated copies of the python interpreter and a set of specific versions of packages (the idea of containerization before and during the earliest years of docker). It is customary to maintain a `requirements.txt` file for each project, which lists the required packages by name and specifies the version, e.g., `wordcloud==1.8.1`.

**We strongly recommend creating environments to avoid global version conflicts.**

That said, nothing prevents you in principle from installing the required packages in the global namespace. We will use the latest version of packages (as listed on https://pypi.org/) and are unlikely to encounter version conflicts. We will not deduce points for using a global installation per se, but note that when version conflicts occur, this falls within your responsibility. When your submission is not executable due to this, we might be unable to award any points for your solution.

### Create a new local Python environment

Start a terminal session and execute the following commands to set up a new virtual environment using [Anaconda](https://docs.anaconda.com/free/anaconda/install/index.html):

1. To create a new environment, run: `conda create --name myenv python=3.9` Replace myenv with the name you wish to give to your virtual environment (e.g. docana).
2. After creating the environment, you need to activate it to use it. Activate the Anaconda environment by executing: `conda activate myenv`. This switches to your newly created environment, allowing you to install packages and run Python isolated from your base environment.
3. Install the IPython kernel package, ipykernel, which is necessary to use Jupyter notebooks within this environment. Execute the following command: `pip install ipykernel`.

**You now have:**

- A new environment called `myenv` using Anaconda.
